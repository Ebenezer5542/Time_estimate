```python
data = nn[['duration','home_pull_value', 'work_pull_value', 'home_pull', 'work_pull',
       'confirm_page', 'town_search', 'house_loc', 'work_search', 'work_loc',
       'transport_mode', 'departure_time', 'arrival_time', 'date',
       'time_taken','home_town_names',
       'work_town_names']]
```

```python
data.drop(columns=['time_taken'], inplace=True)
```

```python
data.info()
```

```python
data['arrival_time']
```

```python
a_time = np.array(data['arrival_time'])
d_time = np.array(data['departure_time'])
```

```python
new_arrival = []
for tx in a_time:
    t = tx[:2] + tx[3:5] + tx[-2:]
    new_arrival.append(t)
    
    
new_departure = []
for tx in d_time:
    d_t = tx[:2] + tx[3:5] + tx[-2:]
    new_departure.append(d_t)
```

```python
len(new_departure)
```

```python
new_departure_sec = []
for ff in new_departure:
    xx = (int(ff[0:2]) *3600) + (int(ff[2:4])*60) + (int(ff[4:]))
    new_departure_sec.append(xx)
    
new_arrival_sec = []
for gg in new_arrival:
    hh = (int(gg[0:2]) *3600) + (int(gg[2:4])*60) + (int(gg[4:]))
    new_arrival_sec.append(hh)
```

```python
time_taken = [ new_arrival_sec[i] - new_departure_sec[i] for i in range(21)]
```

```python
time_taken = np.array(time_taken)
```

```python
time_taken.mean()/3600
```

```python
dept_time_sec = []
for i in data['departure_time']:
    dh = int(i[0:2]) *3600 +(int(i[3:5])*60) + (int(i[6:]))
    dept_time_sec.append(dh)
    
```

```python
data['dept'] = dept_time_sec
```
